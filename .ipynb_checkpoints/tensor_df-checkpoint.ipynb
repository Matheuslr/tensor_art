```python
import requests
import json
import pandas as pd
```

```python
per_page = 250
pagination = 1 
project_count = 0
projects_list =[]
while (per_page * pagination <= project_count):
    url = "https://search.tensor.trade/multi_search?sort_by=statsV2.buyNowPrice:asc&infix=fallback,off,off&page={pagination}&per_page={per_page}&split_join_tokens=off&exhaustive_search=true&prioritize_exact_match=true"
    
    payload = json.dumps({
      "searches": [
        {
          "collection": "nft_collections",
          "q": "*",
          "query_by": "name,slugDisplay,acronym"
        }
      ]
    })
    headers = {
      'x-typesense-api-key': 'DOCUMENT_SEARCH_ONLY_KEY',
      'Content-Type': 'application/json'
    }
    
    response = requests.request("POST", url, headers=headers, data=payload)
    res = response.json()
    projects_list = res["results"][0]["found"]
    
    for item in res["results"][0]["hits"]:
        projects_list.append(item)
    
```

```python
filtered_list = []
for item in projects_list: 
    filtered_list.append(item["document"])
filtered_list
```

```python
df = pd.read_json(json.dumps(filtered_list))
df.to_csv(' mint_list.csv' )
```
